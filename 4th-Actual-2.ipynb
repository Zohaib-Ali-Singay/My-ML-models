# Encoding values

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
```

```python
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2, stratify = y)
```

```python
rfc = RandomForestClassifier()
etc = ExtraTreesClassifier()
abc = AdaBoostClassifier()
gbc = GradientBoostingClassifier()
lgr = LogisticRegression()
gnb = GaussianNB()
svc = SVC()
xgb = XGBClassifier()

models = [rfc, etc, abc, gbc,
         lgr, gnb, svc, xgb]

names = ["Random Forest", "Extra Trees", "Ada Boost",
        "Gradient Boosting", "Logistic Regression", "Naive Bayes",
        "Support Vectors", "XGBoost"]
```

## Training and evaluation

```python
def training(model, name, ax):
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    score = accuracy_score(y_pred, y_test)
    report = classification_report(y_pred, y_test)
    cm = confusion_matrix(y_pred, y_test)
    sns.heatmap(cm, annot=True, ax=ax)
    ax.set_title("{}: {}%".format(name, round(score*100, 2)))
    return score*100, report
```

```python
reports = []
scores = []

fig, axis = plt.subplots(nrows = 2, ncols = 4, figsize = (10, 7))
index = 0
for i in range(2):
    for j in range(4):
        score, report = training(models[index], names[index], axis[i][j])
        scores += [score]
        reports += [report]
        index += 1

plt.tight_layout()
plt.show()
```

```python
dt = pd.DataFrame({"scores" : scores}, index = names)
dt = dt.sort_values("scores", ascending = False)
fig, axis = plt.subplots()
sns.barplot(x = dt.index, y = dt.iloc[:, 0], ax = axis)
for container in axis.containers:
    axis.bar_label(container)
axis.set_xticklabels(axis.get_xticklabels(), rotation=90)
plt.show()
```
