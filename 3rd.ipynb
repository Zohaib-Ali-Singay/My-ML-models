```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators":[10, 20, 40, 50, 70, 100, 120, 130, 150],
             "max_depth" : [10, 20, 30, 40, 50, 60, 70, 80]}
dr = GridSearchCV(RandomForestRegressor(), param_grid = param_grid)
```

```python
df.dtypes
```

```python
function = {"True" : 1, "False" : 0}
for i in list(df.columns):
    if df[i].dtype == 'bool':
        df[i] = [1 if a == True else 0 for a in df[i]]
df.head()
```

```python
x = df.drop(columns = ["price"], axis = 1)
y = df["price"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1)
```

```python
dr.fit(x_train, y_train)
dr.best_params_
```

```python
y_preds = dr.predict(x_test)
mse = mean_squared_error(y_test, y_preds)
mse
```
