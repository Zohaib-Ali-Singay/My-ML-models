```python
def Outlier_iqr(data, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)

    iqr = q3 - q1
    u_l = q3 + 1.5 * iqr
    l_l = q1 - 1.5 * iqr
    return (l_l, u_l)
```

```python
df[df["Number of Ratings"] < 348]["Number of Ratings"].count()
```

```python
l_l, u_l = Outlier_iqr(df, "Number of Ratings")
df.loc[df["Number of Ratings"] > u_l, "Number of Ratings"] = u_l
l_l, u_l = Outlier_iqr(df, "Number of Reviews")
df.loc[df["Number of Reviews"] > u_l,"Number of Reviews"] = u_l
```

```python
df_dummies = pd.get_dummies(df, drop_first = True)
```

```python
df_dummies.head()
```

```python
df_dummies.shape
```

```python
df_dummies = df_dummies.astype("int")
```

```python
df_dummies
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_dummies.drop(columns = ["Price"])
y = df_dummies["Price"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 22)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, lr.predict(X_test))
```

```python
r2_score(y_train, lr.predict(X_train))
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge = Ridge()
params = {"alpha" : [0, 0.1, 0.01, 0.001, 0.05, 1]}
```

```python
from sklearn.model_selection import cross_val_score, GridSearchCV
rg = GridSearchCV(ridge, param_grid = params, scoring  = "r2", cv = 5)
```

```python
rg.fit(X_train, y_train)
```

```python
rg.best_params_
```

```python
best_ridge = rg.best_estimator_
```

```python
r2_score(y_test, best_ridge.predict(X_test))
```

```python
scores = cross_val_score(best_ridge, X_train, y_train, cv = 10)
```

```python
scores.mean()
```

```python
scores
```

Remember always use cross_val_score before the training of model.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr_initial = RandomForestRegressor(min_samples_split = 4, max_depth = 10, n_estimators = 150)
scores = cross_val_score(rfr_initial, X_train, y_train, cv = 5)
```
